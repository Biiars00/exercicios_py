**13. Tendo como dado de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:**

**15. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:**

**a) salário bruto.**

**b) quanto pagou ao INSS.**

**c) quanto pagou ao sindicato.**

**d) o salário líquido.**

**e) calcule os descontos e o salário líquido, conforme a tabela abaixo:**

+ Salário Bruto : R$

- IR (11%) : R$

- INSS (8%) : R$

- Sindicato ( 5%) : R$

- = Salário Liquido : R$

Obs.: Salário Bruto - Descontos = Salário Líquido.

```python
salario_hora = float(input("Quanto você ganha por hora? "))
hora_mes = int(input("Quantas horas você trabalha por mês? "))

salario_bruto = salario_hora * hora_mes
print(f'Seu salário bruto mensal é: {salario_bruto}')

descontos_ir = (11 * salario_bruto) / 100 
descontos_inss = (8 * salario_bruto) / 100
descontos_sindicato = (5 * salario_bruto) / 100

descontos = descontos_ir + descontos_inss + descontos_sindicato
salario_liquido = salario_bruto - descontos
print(f'Seu salário líquido é: R$ {salario_liquido}')

```

**17. Faça um Programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 6 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam 80.00 reais ou em galões de 3.6 litros, que custam 25.00 reais. Informe ao usuário as quantidades de tinta a serem compradas e os respectivos preços em 3 situações:**

**- comprar apenas latas de 18 litros;**

**- comprar apenas galões de 3,6 litros;**

**- misturar latas e galões, de forma que o desperdício de tinta seja menor.**

**- Acrescente 10% de folga e sempre arredonde os valores para cima, isto é, considere latas cheias.** 

```python
import math

metros = float(input("Informe o valor do metro quadrado: "))

litros = metros / 6
latas_galao = litros / 18
valor_galao = 80 * latas_galao

latas = litros / 3.6
valor = 25 * latas

soma_total = latas_galao + latas
valor_total = valor_galao + valor

print(f'Você precisará comprar {math.ceil(latas_galao)} latas de tinta galão (18L). E o valor total é de: R$ {valor_galao:.2f}')
print(f'Você precisará comprar {math.ceil(latas)} latas de tinta (3.6L). E o valor total é de: R$ {valor:.2f}')
print(f'Você precisará comprar {math.ceil(soma_total)} latas de tinta, sendo {math.ceil(latas_galao)} latas de tinta galão e {math.ceil(latas)} latas de 3,6L. E o valor total é de: R$ {valor_total:.2f}')



```

**18. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).**

```python
arquivo = int(input("Informe o tamanho do arquivo (MB): "))
internet = int(input("Informe a velocidade do link de internet (MB): "))

#Tamanho do arquivo multiplicado por 1024 kilobytes, que equivale a 1 MB e dividido por 8(bits). Obs: 1 Byte equivale a 8 bits.
download = (arquivo * 1024 / 8) / internet
#minuto = 60 / 60
#tempo = download / minuto

print(f"O tempo aproximado de download do arquivo é de {download:.0f} minutos.")
```
